```python
import os
import pandas as pd
import numpy as np
df = []
for city in ['az','ba','mj']:
    for file in os.listdir('snps/' + city):
        if not '.csv' in file:
            continue
        df = pd.read_csv('snps/' + city + '/' + file)
        df = df.rename(columns=lambda x: x.strip())
        df = df.reset_index(drop=True)
        df['LOD'] = - np.log10(df['P.value'])
        df = df[['Chromosome','Position','LOD']]
        df['Position_2'] = df['Position'] + 1

        df = df[['Chromosome','Position','Position_2','LOD']]
        df = df.sort_values(['Chromosome','Position'])
        count = 1
        for chromosome in range(1,8):
            for genome in ('A','B','D'):
                df['Chromosome'] = df['Chromosome'].replace(count, str(chromosome) + genome)
                count += 1
        df['Chromosome'] = 'chr' + df['Chromosome'].astype(str)
        df['Position'] = df['Position'].astype(int)
        df['Position_2'] = df['Position_2'].astype(int)
        print(df.head())
        df.to_csv('res/' + file,index=False, header=None, sep=" ")
```
